## Variable Descriptions

This DataFrame contains the following columns:

- `crim`: per capita crime rate by town  
- `zn`: proportion of residential land zoned for lots over 25,000 sq.ft  
- `indus`: proportion of non-retail business acres per town   
- `chas`: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)  
- `nox`: nitrogen oxide concentration (parts per 10 million)   
- `rm`: average number of rooms per dwelling   
- `age`: proportion of owner-occupied units built prior to 1940  
- `dis`: weighted mean of distances to five Boston employment centers   
- `rad`: index of accessibility to radial highways   
- `tax`: full-value property-tax rate per \$10,000   
- `ptratio`: pupil-teacher ratio by town    
- `b`: 1000(Bk - 0.63)^2 where Bk is the proportion of African American individuals by town   
- `lstat`: lower status of the population (percent)   
- `medv`: median value of owner-occupied homes in $10000s

```python
#Goal 1 Load Data
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
df.head()
```

```python
#Goal 2 Use built-in Python functions to explore measures of centrality and dispersion for at least 3 variables
#measures of centrality
crim_mean = df['crim'].mean()
age_mean = df['age'].mean()
tax_median = df['tax'].median()
zn_median = df['zn'].median
#measures of dispersion
crim_stdev = df['crim'].std()
age_stdev = df['age'].std()
tax_variance = df['tax'].var()
zn_variance = df['zn'].var()
df.describe()
```

```python
pd.plotting.scatter_matrix(df, figsize=(10,10));
```

```python
df['dis'].hist();
plt.title('Mean Distance from 5 Boston Employment Centers')
plt.xlabel('Mean Distance from Employment Centers')
plt.ylabel('Number of Houses')
    
```

```python
#Right skewed distribution.  Most houses are close to employment centers
```

```python
df['rm'].hist();
plt.title('Distribution of Rooms per Household')
plt.xlabel('Number of Rooms')
plt.ylabel('Number of Houses');
```

```python
#Very symmetric with the typical number of rooms being around 6 rooms per household
```

```python
df['medv'].hist();
plt.title('Distribution of Median Value of Homes')
plt.xlabel('Median Home Value for households over time')
plt.ylabel('Number of Houses');
```

```python
#right skewed again with most median values being around $200,000 and a few outliers that are
#above $500,000
```

```python
df['crim'].hist();
plt.title('Crime Rate per Capita')
```

```python
#Most crime rates are well below 10
df['crim'].quantile(.5)
#The median crime rate is .26
```

```python
import numpy as np
```

```python
for n, feat1 in enumerate(df.columns):
    for feat2 in df.columns[n+1:]:
        r2 = np.corrcoef(df[feat1], df[feat2])[0][1]
        if r2 > .7:
            print('{} and {} have a correlation coefficient of {}!'.format(feat1, feat2, r2))
```

```python
df['decades'] = df.age//10
to_plot = df.groupby('decades').medv.mean()
to_plot.plot(kind='barh', figsize=(10,8))
plt.ylabel('House Age in Decades')
plt.xlabel('Average Median Value of Homes')
plt.title('Average Home Values by Home Age');
```
